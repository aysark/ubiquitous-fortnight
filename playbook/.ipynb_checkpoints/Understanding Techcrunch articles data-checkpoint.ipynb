```python
import json
from urllib.request import Request, urlopen
import pandas as pd

request=Request('https://public-api.wordpress.com/wp/v2/sites/techcrunch.com/posts?per_page=100')
response = urlopen(request).read().decode('utf8')
print(response)
data = json.loads(response)
```

```python
df = pd.io.json.json_normalize(data)

print(df.shape)
print('-'*40)
df.info()
print('-'*40)
df.describe()
```
